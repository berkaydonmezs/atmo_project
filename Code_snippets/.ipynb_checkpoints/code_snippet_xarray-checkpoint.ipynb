# How to read gridded MCH monthly data

# Introduction
You can either read these data with netcdf or xarray. Here, we give you the basic lines to read it with xarray. 
If you prefer working with netcdf, you can look at the code_snippet.py provided in the same enacshare.

Here are only some basic data analyses shown to give you a good starting basis.

```python
# You will need these packages:
import os, sys, glob

import numpy as np 
import xarray as xr 

# To get the right datetime object
import datetime
from dateutil.relativedelta import relativedelta

# For plotting outside of the xarray package
import matplotlib.pyplot as plt
```

```python
# Here is where you define the location of your files:
dir_path = '' # set the path to your data here
product = "RhiresM"
filepaths = glob.glob(os.path.join(dir_path, product, "*.nc"))
```

# 1) How to read the data

```python
# This cell does the major work: it reads all the available files that are in the dir_path:
ds = xr.open_mfdataset(filepaths, engine="netcdf4", chunks={}, decode_times=False)
ds
```

In the output of the above cell, you should see the dimensions, coordinates, data variables and attributes of the read data.
To get a better idea on what the time contains, run the following command:

```python
ds.time
```

As you can see in the output (Attributes: units:), the time is in "months since 1900-01-01 00:00:00". However, we would like to work with these months, so we use the following lines to create a date-time object.

```python
start_time_str = ds["time"].attrs["units"].replace("months since ", "")
start_time_obj = datetime.datetime.strptime(start_time_str, '%Y-%m-%d %H:%M:%S')

# In the following, we add the relative timedelta in months to our starting time. These will be stored in ds['time'].
timesteps = [start_time_obj + relativedelta(months=im) for im in ds['time']]
ds['time'] = timesteps
```

# 2) How to plot the data

```python
# We can plot the data using the built-in method in the xarray package:

# The variable name within the data corresponds to the product name
variable = product 

# We are plotting the first timestamp within ds[variable][time, E, N] (see above for the indices)
ds[variable][0,:,:].plot()
```

# 3) Basic data analyses

## Average and sums over time

```python
# Average the data over time
clim = ds[variable].mean(dim='time')

# Setup a figure and subplots (optional)
fig = plt.figure()
ax = plt.subplot2grid((1,1),(0,0))
clim.plot.contourf(ax=ax, xcmap='viridis', add_colorbar=True, title='Climatological mean (all years)')
ax.set_title('Climatological mean (Jan 1961- Dec 2022)')
```

```python
year1961 = ds[variable].sel(time=slice('1961-01','1961-12')).sum(dim='time')
year2022 = ds[variable].sel(time=slice('2022-01','2022-12')).sum(dim='time')

# Irregular levels to illustrate the use of a proportional colorbar
levels = np.arange(300,3000, 300)

# Setup figure:
fig = plt.figure(figsize=(12,5))
ax1 = plt.subplot2grid((1,2),(0,0))
ax2= plt.subplot2grid((1,2),(0,1))

# Plot data
year1961.plot.contourf(ax=ax1, xcmap='viridis', add_colorbar=True, levels=levels)
ax1.set_title('Precipitation sum 1961')

year2022.plot.contourf(ax=ax2, xcmap='viridis', add_colorbar=True, levels=levels)
ax2.set_title('Precipitation sum 2022')

# Arrange subplots
plt.tight_layout()
```

## Select years and average over space

```python
# Get all the years, resample them by years by taking the means
var_year =  ds[variable].sel(time=slice('1980-01',None)).resample(time='Y').mean(dim='time')

# Take the average over space and plot in time
var_year.mean(dim=('N', 'E')).plot()
```

## Separate the data into seasons

```python
# Group by seasons and plot maps
ds[variable].groupby('time.season').mean(dim='time').plot(col='season', col_wrap=2)
```
